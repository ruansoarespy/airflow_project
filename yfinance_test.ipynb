```python
def acoes(ticker_symbol, start_date, end_date, interval="1h"):

    ticker = yf.Ticker(ticker_symbol)
    hist = ticker.history(
        interval=interval,
        start=start_date,
        end=end_date,
        prepost=True
    )
    return hist
```
